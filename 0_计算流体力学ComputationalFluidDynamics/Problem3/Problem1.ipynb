# Problem

编程作业1.采用偏微分方程数值解法生成NACA0012翼型O型(C型)网格，并采用速度势方程求解绕NACA0012翼型迎角为零度时的低速不可压无粘流场。

## NACA0012翼型（修型）表面坐标

$$
y_{u, l}=\pm 0.6\left[-0.1015 x^{4}+0.2843 x^{3}-0.3576 x^{2}-0.1221 x\right. +0.2969 \sqrt{x}]
$$

## 用于网格生成的椭圆型偏微分方程

$$
\frac{\partial^{2} \xi}{\partial x^{2}}+\frac{\partial^{2} \xi}{\partial y^{2}}=0\\
\frac{\partial^{2} \eta}{\partial x^{2}}+\frac{\partial^{2} \eta}{\partial y^{2}}=0\\
\alpha \frac{\partial^{2} x}{\partial \xi^{2}}-2 \beta \frac{\partial^{2} x}{\partial \xi \partial \eta}+\gamma \frac{\partial^{2} x}{\partial \eta^{2}}=0 \\
 \alpha \frac{\partial^{2} y}{\partial \xi^{2}}-2 \beta \frac{\partial^{2} y}{\partial \xi \partial \eta}+\gamma \frac{\partial^{2} y}{\partial \eta^{2}}=0 \\
 \alpha=x_{\eta}^{2}+y_{\eta}^{2}, \quad \beta=x_{\xi} x_{\eta}+y_{\xi} y_{\eta}, \gamma=x_{\xi}^{2}+y_{\xi}^{2} 
$$

# 求解步骤

## 边界条件设定

### O型网格

* 给定远场点坐标，如30倍弦长以上的圆上各点
* 给定翼型表面点坐标，表面点分布原则是保证流动变化剧烈的区域分布密，变化平缓的区域分布稀疏；

例如按如下关系式分布：


$$
\left. \begin{array}  { l  }  { x = \frac { c } { 2 } ( 1 + \cos \theta ) } \\ { \theta = [ 0 , 2 \pi ] } \end{array} \right.
$$

## 给定初始值

根据给定边界值插值出每一网格点的初始值

$$
x _ { i , j } ( 0 ) , y _ { i , j } ( 0 )
$$

## 迭代求解

根据差分方程进行迭代求解，注意应用边界条件直到达到收敛标准。

* 历遍所有i，j：
    $$
    \left. \begin{array}  { l  }  {  | x _ { i , j } ^ { ( n + 1 ) } - x _ { i , j } ^ { ( n ) } | \lt e } \\ {  | y _ { i , j } ^ { ( n + 1 ) } - y _ { i , j } ^ { ( n ) } | \lt e } \end{array} \right.
    $$
    
* 差分方程：
    $$
    \alpha \frac{x_{i+1, j}-2 x_{i, j}+x_{i-1, j}}{(\Delta \xi)^{2}}-2 \beta \frac{x_{i+1, j+1}+x_{i-1, j-1}-x_{i+1, j-1}-x_{i-1, j+1}}{4 \Delta \xi \Delta \eta} +\gamma \frac{x_{i, j+1}-2 x_{i, j}+x_{i, j-1}}{(\Delta \eta)^{2}}=0\\
    \alpha \frac{y_{i+1, j}-2 y_{i, j}+y_{i-1, j}}{(\Delta \xi)^{2}}-2 \beta \frac{y_{i+1, j+1}+y_{i-1, j-1}-y_{i+1, j-1}-y_{i-1, j+1}}{4 \Delta \xi \Delta \eta} +\gamma \frac{y_{i, j+1}-2 y_{i, j}+y_{i, j-1}}{(\Delta \eta)^{2}}=0
    $$
    其中：
    $$
    \alpha=x_{\eta}^{2}+y_{\eta}^{2}=\left(\frac{x_{i, j+1}-x_{i, j-1}}{2 \Delta \eta}\right)^{2}+\left(\frac{y_{i, j+1}-y_{i, j-1}}{2 \Delta \eta}\right)^{2}\\
    \beta=x_{\xi} x_{\eta}+y_{\xi} y_{\eta}=\frac{x_{i+1, j}-x_{i-1, j}}{2 \Delta \xi} \cdot \frac{x_{i, j+1}-x_{i, j-1}}{2 \Delta \eta}+\frac{y_{i+1, j}-y_{i-1, j}}{2 \Delta \xi} \cdot \frac{y_{i, j+1}-y_{i, j-1}}{2 \Delta \eta}\\
    \gamma=x_{\xi}^{2}+y_{\xi}^{2}=\left(\frac{x_{i+1, j}-x_{i-1, j}}{2 \Delta \xi}\right)^{2}+\left(\frac{y_{i+1, j}-y_{i-1, j}}{2 \Delta \xi}\right)^{2}
    $$

* 迭代格式

    * 简单迭代
        $$
        \begin{aligned} x_{i, j}=\frac{1}{2}[& \alpha\left(x_{i+1, j}+x_{i-1, j}\right)+\gamma\left(x_{i, j+1}+x_{i, j-1}\right) \\ &\left.\quad-\frac{\beta}{2}\left(x_{i+1, j+1}+x_{i-1, j-1}-x_{i+1, j-1}-x_{i-1, j+1}\right)\right] /(\alpha+\gamma) \end{aligned}\\
        \begin{aligned} y_{i, j}=\frac{1}{2}[& \alpha\left(y_{i+1, j}+y_{i-1, j}\right)+\gamma\left(y_{i, j+1}+y_{i, j-1}\right) \\ &\left.\quad-\frac{\beta}{2}\left(y_{i+1, j+1}+y_{i-1, j-1}-y_{i+1, j-1}-y_{i-1, j+1}\right)\right] /(\alpha+\gamma) \end{aligned}
        $$

* 边界条件的应用

    * O型网格：割缝AI(CD)为周期性边界条件，物面ABC和远场IHGFED坐标值是给定的。



    * C型网格：割缝AI(CD)为周期性边界条件，物面ABC和远场LJIHGFED坐标值是给定的。



## 求解流场

* 全速势方程：

$$
\frac { \partial ^ { 2 } \phi } { \partial x ^ { 2 } } + \frac { \partial ^ { 2 } \phi } { \partial y ^ { 2 } } = 0
$$

​		转换到计算平面：
$$
\alpha \frac{\partial^{2} \phi}{\partial \xi^{2}}+\gamma \frac{\partial^{2} \phi}{\partial \eta^{2}}-2 \beta \frac{\partial^{2} \phi}{\partial \xi \partial \eta}=0 
$$

* 边界条件：

    * 远场自由来流条件

    $$
    \phi = V _ { \infty } x
    $$

    * 物面条件
        $$
        \frac{\partial \phi}{\partial n}=0 \Rightarrow \frac{\partial \phi}{\partial \eta}=\frac{x_{\xi} x_{\eta}+y_{\xi} y_{\eta}}{x_{\xi}^{2}+y_{\xi}^{2}} \frac{\partial \phi}{\partial \xi}
        $$

* 物理参数
    $$
    J \equiv \frac{\partial(x, y)}{\partial(\xi, \eta)}=\left|\begin{array}{cc}\frac{\partial x}{\partial \xi} & \frac{\partial y}{\partial \xi} \\ \frac{\partial x}{\partial \eta} & \frac{\partial y}{\partial \eta}\end{array}\right|=\frac{\partial x}{\partial \xi} \frac{\partial y}{\partial \eta}-\frac{\partial x}{\partial \eta} \frac{\partial y}{\partial \xi}
    $$
